# Sales Pipeline Analysis

## Python Data Analytics Project

This project analyzes a sales pipeline dataset to evaluate pipeline health, revenue concentration, sales performance, partner influence, and revenue forecasting.

**Tools:** Python, Pandas, NumPy, Matplotlib, Seaborn


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
pd.set_option("display.max_columns", None)
```

## Load Data

```
df = pd.read_csv('../data/processed/sales_pipeline_clean.csv')
df.head()
```

## Dataset Overview

```
print(df.shape)
df.info()
df.describe()
df.isnull().sum()
```

## Pipeline Stage Analysis

```
stage_value = (
    df.groupby("stage")["deal_value"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(10,6))
sns.barplot(x=stage_value.values, y=stage_value.index)
plt.title("Pipeline Value by Stage")
plt.tight_layout()
plt.savefig("../visualizations/pipeline_by_stage.png")
plt.show()
```

## Opportunity Size Analysis

```
size_analysis = (
    df.groupby("size_bucket")["deal_value"]
      .sum()
      .reset_index()
)

plt.figure(figsize=(10,6))
sns.barplot(data=size_analysis, x="size_bucket", y="deal_value")
plt.title("Revenue by Opportunity Size")
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig("../visualizations/opportunity_size_distribution.png")
plt.show()
```

## Sales Representative Performance

```
rep_performance = (
    df.groupby("sales_rep")
      .agg(total_revenue=("deal_value","sum"),
           opportunities=("opportunity_id","count"))
      .sort_values("total_revenue", ascending=False)
)

plt.figure(figsize=(12,7))
sns.barplot(x=rep_performance["total_revenue"], y=rep_performance.index)
plt.title("Sales Representative Performance")
plt.tight_layout()
plt.savefig("../visualizations/sales_rep_performance.png")
plt.show()
```

## Partner Analysis

```
partner_analysis = (
    df.groupby("partner_involved")["deal_value"]
      .sum()
      .reset_index()
)

plt.figure(figsize=(8,5))
sns.barplot(data=partner_analysis, x="partner_involved", y="deal_value")
plt.title("Partner Influence on Revenue")
plt.tight_layout()
plt.savefig("../visualizations/partner_involvement.png")
plt.show()
```

## New vs Existing Accounts

```
account_analysis = (
    df.groupby("account_type")["deal_value"]
      .sum()
      .reset_index()
)

plt.figure(figsize=(8,5))
sns.barplot(data=account_analysis, x="account_type", y="deal_value")
plt.title("Revenue by Account Type")
plt.tight_layout()
plt.savefig("../visualizations/account_breakdown.png")
plt.show()
```

## Product Category Analysis

```
product_analysis = (
    df.groupby("product_category")["deal_value"]
      .sum()
      .reset_index()
)

plt.figure(figsize=(8,5))
sns.barplot(data=product_analysis, x="product_category", y="deal_value")
plt.title("Revenue by Product Category")
plt.tight_layout()
plt.savefig("../visualizations/product_category_analysis.png")
plt.show()
```

## Revenue Forecast

```
forecast = (
    df.groupby("forecast_month")["deal_value"]
      .sum()
      .reset_index()
)

plt.figure(figsize=(12,6))
plt.plot(forecast["forecast_month"], forecast["deal_value"])
plt.xticks(rotation=45)
plt.title("Forecasted Revenue")
plt.tight_layout()
plt.savefig("../visualizations/revenue_forecast.png")
plt.show()
```

## Pipeline Health

```
next_step = df["next_step_defined"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(next_step.values, labels=next_step.index, autopct="%1.1f%%")
plt.title("Next Step Status")
plt.savefig("../visualizations/next_step_status.png")
plt.show()
```

## Executive Summary

### Key Findings

- Pipeline value is heavily concentrated among large opportunities.
- Partner-assisted opportunities contribute a significant share of revenue.
- A large percentage of opportunities lack defined next steps.
- Revenue generation varies significantly across sales representatives.
- Expected revenue is concentrated in a limited number of future periods.
